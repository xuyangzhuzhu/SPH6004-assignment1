```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('train_preprocessed-nonPCA.csv',index_col=[0])
```

```python
df1 = pd.read_csv('test_preprocessed-nonPCA.csv',index_col=[0])
df1
```

```python
df
```

```python
akitype = df['aki'].astype( "category" )
akitype = akitype.value_counts()
akitype
```

```python
akitype = df1['aki'].astype( "category" )
akitype = akitype.value_counts()
akitype
```

```python
import numpy as np

# Create a numpy array for the categories 将列的数据类型转换成“类别”(category)类型
x = np.array( akitype.index.values )

# Convert the count table into a numpy array
y = akitype.to_numpy()

plt.bar( x, y )
plt.xticks([0, 1, 2, 3])
plt.xlabel( "Aki Type" )
plt.ylabel( "Number of occurances" )
plt.title('Aki categories for traning dataset')
plt.show()
```
